
## Explore presidential speeches  (Inaugural speeches)

#### We will look at the inaugural speeches of the last three Presidents of the United States, i.e., Bush, Obama and Trump. There are several ways to examine the most frequent words used in a text. We will use wordclouds to visualize the most frequent words used in the speeches.

#### In order to further explore the context of the speeches, we use collocation to see the most frequent bigrams in the speeches.



```python
## Initialization : load needed libraries

import nltk
import pandas as pd
import matplotlib.pyplot as plt
import numpy 

from wordcloud import WordCloud, STOPWORDS
from PIL import Image

from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
from string import punctuation


## Get inaugural speeches of Bush, Obama and Trump from nltk corpus 

bush01 = nltk.corpus.inaugural.words('2001-Bush.txt')
bush05 = nltk.corpus.inaugural.words('2005-Bush.txt')
obama09 = nltk.corpus.inaugural.words('2009-Obama.txt')
obama13 = nltk.corpus.inaugural.words('2013-Obama.txt')
trump17 = nltk.corpus.inaugural.words('2017-Trump.txt')
```

#### Remove stop words and punctuation before generating word clouds, using stopwords from nltk.  

```python
## Remove stop words and punctuations from the speech texts

bush01_rsw = [token.lower() for token in bush01 if token.lower() not in stopWords and token not in punctuation]
bush05_rsw = [token.lower() for token in bush05 if token.lower() not in stopWords and token not in punctuation]

obama09_rsw = [token.lower() for token in obama09 if token.lower() not in stopWords and token not in punctuation]
obama13_rsw = [token.lower() for token in obama13 if token.lower() not in stopWords and token not in punctuation 
               and token.lower() != '\x80\x94'] 

trump17_rsw = [token.lower() for token in trump17 if token.lower() not in stopWords and token not in punctuation]
```

#### Generate wordclouds and masks for the wordclouds.

```python
## Wordclouds for Bush
##
BushMask = numpy.array(Image.open('C:/Users/meiye/Venv1/images/bush2.png')) # Bush's image as wordcloud mask

wcloud_b1 = WordCloud(width = 920, height = 800, # Generate wordcloud for Bush 2001 speech
                background_color ='black',
                mask = BushMask,
                #stopwords = STOPWORDS, 
                max_words = 100,
                min_font_size = 6).generate(str(bush01_rsw)) 

wcloud_b2 = WordCloud(width = 920, height = 800, # Generate wordcloud for Bush 2005 speech
                background_color ='black',
                mask = BushMask,
                #stopwords = STOPWORDS, 
                max_words = 100,
                min_font_size = 6).generate(str(bush05_rsw)) 

## Obama wordclouds
##
mask = numpy.array(Image.open('C:/Users/meiye/Venv1/images/obama2.png'))  # Obama's image as wordcloud mask

wcloud_o1= WordCloud(width = 810, height = 800, # Generate the wordclouds for Obama 2009 speech
                background_color ='black',
                mask = mask,
#                 stopwords = stopWords, 
                max_words = 100,
                min_font_size = 6).generate(str(obama09_rsw)) 

wcloud_o2= WordCloud(width = 810, height = 800, # Generate the wordclouds for Obama 2013 speech
                background_color ='black',
                mask = mask,
#                 stopwords = stopWords, 
                max_words = 100,
                min_font_size = 6).generate(str(obama13_rsw)) 

## Trump wordcloud
##
T_mask = numpy.array(Image.open('C:/Users/meiye/Venv1/images/trump1.png'))  # Trumps's image as wordcloud mask

wcloud_t= WordCloud(width = 850, height = 700,   # Generate the wordcloud for Trump 2017 speech
                background_color ='black',
                mask =T_mask,
#                 stopwords = S*TOPWORDS, #stopWords, 
                max_words = 100,
                min_font_size = 6).generate(str(trump17_rsw)) 
```

```python
## Plot Bush wordclouds
#
plt.figure(figsize = (12, 12))
plt.subplot(2,2,1)  
plt.title("Bush 2001")
plt.imshow(wcloud_b1, interpolation="bilinear") 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.subplot(2,2,2)
plt.title("Bush 2005")
plt.imshow(wcloud_b2, interpolation="bilinear") 
plt.axis("off") 
plt.tight_layout(pad = 0) 
```

In his 2001 inaugural speech, Bush's most frequent words were *$\color{navy}{nation}$*, *$\color{navy}{us}$*, *$\color{navy}{story}$*, *$\color{navy}{citizen}$*, *$\color{navy}{country}$* and *$\color{navy}{america}$*. Words such as *$\color{navy}{common}$*, *$\color{navy}{new}$*, *$\color{navy}{freedom}$*, *$\color{navy}{promise}$*, *$\color{navy}{americans}$* and *$\color{navy}{courage}$* were mentioned 5 times each (see frequency plot below). 

In 2005, his speech emphasized words like *$\color{navy}{freedom}$*, *$\color{navy}{america}$* and *$\color{navy}{liberty}$*. These words were used at least 15 times and *$\color{black}{liberty}$* was not very frequent (only once) in his previous speech. Words that were mentioned more often in his second speech than the first are *$\color{navy}{history}$*, *$\color{navy}{free}$*, *$\color{navy}{justice}$*, *$\color{navy}{human}$* and *$\color{navy}{hope}$*. The word *$\color{black}{human}$* did not appear in his previous speech. The context which *$\color{black}{human}$* was used are shown below. 

```python
nltk.Text(bush05).concordance('human', width = 50)
```

```python
fdist_bush01 = nltk.FreqDist(bush01_rsw)
fdist_bush05 = nltk.FreqDist(bush05_rsw)

plt.figure(figsize=(10,3))
plt.title('Bush 2001')
plt.tight_layout()
fdist_bush01.plot(20)

plt.figure(figsize=(10,3))
plt.title('Bush 2005')
plt.tight_layout()
fdist_bush05.plot(20)

plt.show()
```

```python
# plot Obama's WordClouds 
#
plt.figure(figsize = (12, 12))
plt.subplot(2,2,1)
plt.title("Obama 2009")
plt.imshow(wcloud_o1, interpolation="gaussian") 
plt.axis("off") 
plt.tight_layout(pad = 2) 

plt.subplot(2,2,2)
plt.title("Obama 2013")
plt.imshow(wcloud_o2, interpolation="gaussian") 
plt.axis("off") 
plt.tight_layout(pad = 2) 
```

In both his speeches, Obama's frequent words were rather similar, such as *$\color{navy}{us}$*, *$\color{navy}{nation}$*, *$\color{navy}{people}$*, *$\color{navy}{America}$*, *$\color{navy}{must}$*, *$\color{navy}{generation}$* and *$\color{navy}{new}$*.

However, *$\color{navy}{world}$* is mentioned more often in his 2009 speech and *$\color{navy}{freedom}$* is used more often in 2013. Below are the contexts which these words were used.


```python
print('Concordance of "world" in 2009 speech : \n') 
nltk.Text(obama09).concordance('world', width=60)

print('\n\nConcordance of "freedom" in 2013 speech : \n ') 
nltk.Text(obama13).concordance('freedom', width=60)
```

```python
fdist_obama09 = nltk.FreqDist(obama09_rsw)
fdist_obama13 = nltk.FreqDist(obama13_rsw)

plt.figure(figsize=(10,3))
plt.title('Obama 2009')
plt.tight_layout()
fdist_obama09.plot(30)

plt.figure(figsize=(10,3))
plt.title('Obama 2013')
plt.tight_layout()
fdist_obama13.plot(30)

plt.show()
```

```python
# plot Trupm's WordCloud 
#
plt.figure(figsize = (6, 5))

plt.title("Trump 2017")
plt.imshow(wcloud_t, interpolation="bilinear") 
plt.axis("off") 
plt.tight_layout(pad = -1) 

plt.show() 
```

In Trump's 2017 speech, the most frequent words are *$\color{navy}{america}$* and *$\color{navy}{american}$* follow by *$\color{navy}{people}$*, *$\color{navy}{country}$*, *$\color{navy}{nation}$*, *$\color{navy}{world}$* which are pretty typical of inaugural speeches (at least compared to the last two presidents). Trump also used the word *$\color{navy}{great}$* a lot compared to the previous two presidents. 


```python
fdist_trump17 = nltk.FreqDist(trump17_rsw)

plt.figure(figsize=(10,3))
plt.title('Trump 2017')
plt.tight_layout()
fdist_trump17.plot(30)

plt.show()

```

```python
print('\n\nTrump 2017')
nltk.Text(trump17).concordance('great')

print('\n\nObama 2013')
nltk.Text(obama13).concordance('great')

print('\n\nBush 2001')
nltk.Text(bush01).concordance('great')
```


#### We find that frequent bigrams give a whole lot more context than examining frequent words. In the following are the frequent bigrams of the speeches of the three presidents. 





```python
bush01_text = nltk.Text(bush01)
bush05_text = nltk.Text(bush05)

print('Bush 2001 : ', '; '.join(bush01_text.collocation_list()))
print('Bush 2005 : ', '; '.join(bush05_text.collocation_list()))
```

```python
obama09_text = nltk.Text(obama09)
obama13_text = nltk.Text(obama13)

print('Obama 2009 : ' ,'; '.join(obama09_text.collocation_list()))
print('Obama 2013 : ' , '; '.join(obama13_text.collocation_list()))
```

```python
trump17_text = nltk.Text(trump17)

print('; '.join(trump17_text.collocation_list()))
```
